```python
import numpy as np
import pandas as pd

# Tạo danh sách các chữ cái từ 'a' đến 'z'
mylist = list('abcdefghijklmnopqrstuvwxyz')

# Tạo mảng số nguyên từ 0 đến 25 bằng NumPy
myarr = np.arange(26)

# Tạo dictionary từ mylist và myarr
mydict = dict(zip(mylist, myarr))

# Chuyển đổi dictionary thành DataFrame
df = pd.DataFrame(list(mydict.items()), columns=['Letter', 'Number'])

# Hiển thị DataFrame
print(df)
```

## Bài 4.17

```python
# Series đầu vào
ser = pd.Series(['2010-01-01', '02-02-2011', '20120307', '2013/04/06', '2014-05-07', '2015-06-08T01:02'])

# Chuyển đổi chuỗi ngày thành chuỗi thời gian với các định dạng khác nhau
formats = ['%Y-%m-%d', '%d-%m-%Y', '%Y%m%d', '%Y/%m/%d', '%Y-%m-%d', '%Y-%m-%dT%H:%M']

# Hàm để chuyển đổi từng chuỗi ngày
def parse_date(date_str):
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except (ValueError, TypeError):
            continue
    return pd.NaT

# Áp dụng hàm chuyển đổi cho Series
time_series = ser.map(parse_date)

print("Chuỗi thời gian:")
print(time_series)
```

```python
import pandas as pd

# a) Đọc file stock1.csv và thực hiện các yêu cầu
stock1 = pd.read_csv(r'D:\Python_nâng_cao\17A3\BTVN_C4\stocks1.csv',encoding='utf-8')

# Hiển thị 5 dòng đầu và cuối của stock1
print("5 dòng đầu của stock1:")
print(stock1.head())
print("\n5 dòng cuối của stock1:")
print(stock1.tail())

# Kiểu dữ liệu (dtype) của các cột trong stock1
print("\nKiểu dữ liệu (dtype) của các cột trong stock1:")
print(stock1.dtypes)

# Thông tin (info) của stock1
print("\nThông tin của stock1:")
print(stock1.info())

# b) Đọc file stock2.csv và thực hiện các yêu cầu
stock2 = pd.read_csv(r'D:\Python_nâng_cao\17A3\BTVN_C4\stocks2.csv',encoding='utf-8')

# Hiển thị 5 dòng đầu và cuối của stock2
print("\n5 dòng đầu của stock2:")
print(stock2.head())
print("\n5 dòng cuối của stock2:")
print(stock2.tail())

# Kiểu dữ liệu (dtype) của các cột trong stock2
print("\nKiểu dữ liệu (dtype) của các cột trong stock2:")
print(stock2.dtypes)

# Thông tin (info) của stock2
print("\nThông tin của stock2:")
print(stock2.info())

# c) Đọc file companies.csv và thực hiện các yêu cầu
companies = pd.read_csv(r'D:\Python_nâng_cao\17A3\BTVN_C4\companies.csv',encoding='utf-8')

# Xem dữ liệu của companies
print("\nDữ liệu của companies:")
print(companies.head())

# Kiểu dữ liệu (dtype) của các cột trong companies
print("\nKiểu dữ liệu (dtype) của các cột trong companies:")
print(companies.dtypes)

# Thông tin (info) của companies
print("\nThông tin của companies:")
print(companies.info())

# Kiểm tra xem có dữ liệu Null trong stock1 hay không
print("\nKiểm tra dữ liệu Null trong stock1:")
print(stock1.isnull().sum())

# Nếu có Null, thay thế theo quy tắc đã mô tả
if stock1.isnull().values.any():
    stock1['high'] = stock1['high'].fillna(stock1.groupby('symbol')['high'].transform('max'))
    stock1['low'] = stock1['low'].fillna(stock1.groupby('symbol')['low'].transform('min'))

# Tạo dataframe stocks bằng cách gộp stock1 và stock2 theo dòng
stocks = pd.concat([stock1, stock2], ignore_index=True)

# Xem 15 dòng dữ liệu cuối của stocks
print("\n15 dòng cuối của stocks:")
print(stocks.tail(15))

# Tạo dataframe stocks_companies bằng cách gộp stocks và companies
stocks_companies = pd.merge(stocks, companies, left_on='symbol', right_on='name', how='left')

# Xem 5 dòng đầu của stocks_companies
print("\n5 dòng đầu của stocks_companies:")
print(stocks_companies.head())

# Tính toán giá (open, high, low, close) trung bình và volume trung bình của mỗi công ty
avg_prices_volume = stocks_companies.groupby('name')[['open', 'high', 'low', 'close', 'volume']].mean()

# Giá đóng cửa (close) trung bình, lớn nhất và nhỏ nhất ở mỗi công ty
close_stats = stocks_companies.groupby('name')['close'].agg(['mean', 'max', 'min'])

print("\nGiá (open, high, low, close) trung bình và volume trung bình của mỗi công ty:")
print(avg_prices_volume)

print("\nGiá đóng cửa (close) trung bình, lớn nhất và nhỏ nhất ở mỗi công ty:")
print(close_stats)

# Tạo cột parsed_time trong stocks_companies bằng cách chuyển thời gian sang định dạng DateTime
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])

# Kiểu dữ liệu của cột parsed_time
print("\nKiểu dữ liệu của cột parsed_time:")
print(stocks_companies['parsed_time'].dtype)

# Hiển thị 5 dòng đầu của stocks_companies
print("\n5 dòng đầu của stocks_companies:")
print(stocks_companies.head())

# Thêm cột result, nếu giá 'close' > 'open' thì 'up', ngược lại 'down'
stocks_companies['result'] = stocks_companies.apply(lambda row: 'up' if row['close'] > row['open'] else 'down', axis=1)

# Hiển thị kết quả cuối cùng
print("\nDữ liệu cuối cùng của stocks_companies:")
print(stocks_companies.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu vào Pandas DataFrame
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
links = pd.read_csv('links.csv')

# 2. Xử lý dữ liệu bị thiếu (missing data)
# Kiểm tra xem có dữ liệu missing hay không
print(movies.isnull().sum())
print(ratings.isnull().sum())

# Nếu có dữ liệu bị thiếu, có thể thay thế hoặc loại bỏ:
movies.dropna(subset=['title', 'genres'], inplace=True)  # Loại bỏ dòng thiếu dữ liệu quan trọng
ratings.fillna(0, inplace=True)  # Thay thế NaN trong ratings bằng 0

# 3. Gộp DataFrame: Gộp movies và ratings theo movieId
movie_ratings = pd.merge(ratings, movies, on='movieId', how='inner')

# 4. Lọc dữ liệu
# Lọc ra các bộ phim có xếp hạng trung bình lớn hơn 4
high_rated_movies = movie_ratings.groupby('title').agg({'rating': 'mean'}).reset_index()
high_rated_movies = high_rated_movies[high_rated_movies['rating'] > 4]
print(high_rated_movies.head())

# 5. Thống kê dữ liệu
# Đếm số lượng bộ phim theo thể loại
genres_count = movies['genres'].value_counts()
print(genres_count)

# 6. Parsing Timestamps: Chuyển đổi cột timestamp thành datetime
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')

# 7. Trực quan hóa dữ liệu
# Histogram xếp hạng ratings
plt.figure(figsize=(10,6))
sns.histplot(ratings['rating'], kde=True, bins=10)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Trực quan hóa phân bố số lượng phim theo thể loại
plt.figure(figsize=(12,8))
sns.barplot(x=genres_count.index, y=genres_count.values)
plt.title('Number of Movies per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)
plt.show()

# Phân phối xếp hạng trung bình theo thể loại
genre_ratings = movie_ratings.groupby('genres').agg({'rating': 'mean'}).reset_index()
plt.figure(figsize=(12,8))
sns.barplot(x='rating', y='genres', data=genre_ratings)
plt.title('Average Rating per Genre')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.show()
```
